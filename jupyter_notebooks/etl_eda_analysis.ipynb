```python
import pandas as pd 
from pathlib import Path 
# Resolve the absolute path safely 

file_Path= Path ("..") / ("Data") / "pollution_us_2000_2016.csv"
assert file_Path.exists(), f"CSV not found at {file_Path.resolve()}"

out_Path = Path("outputs") / "pollution_use_200_2016.csv"
out_Path.parent.mkdir(parents=True, exist_ok=True) # ensure outputs/ exists
if out_Path.exists():
    out_Path.unlink() # start fresh 

chunks= pd.read_csv(file_Path, chunksize=200_000) # adjust if needed 
first= True

for i, chunk in enumerate (chunks):
    print (f"Processing chunk {i+1}...")
# apply transformations on each chunk 
chunk ["Date Local"]= pd.to_datetime(chunk["Date Local"], errors= "coerce")
chunk ["Year"]= chunk["Date Local"].dt.year
chunk ["Month"]= chunk["Date Local"].dt.month
chunk ["Quarter"]= chunk["Date Local"].dt.quarter

# (optional) drop rename here if you want to keep everything streaming:
# drop_cols= [c for c in chuck.columns if "unnamed" in c or "1st Max Hour" in c or "Unit" in c]
# chunk= chunk.drop (columns = drop_cols, errors="ignore")

# Append to output incrementally 
chunk.to_csv(out_Path, mode="a", index=False, header=first)
first = False 
print (f" Finished.wrote cleaned data to: {out_Path.resolve()}")

```
